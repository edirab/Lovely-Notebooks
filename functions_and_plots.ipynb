```python
from matplotlib import pyplot as plt
import matplotlib as mpl
from pprint import pprint
from math import exp, sin
import numpy as np
```

```python
x = list()
y = list()

a = -5
b = 10
step = 0.1

for i in range(int((b-a)/step)):
    x1 = a + i * step
    y1 = exp(-x1) + 10*sin(x1)
    
    x.append(x1)
    y.append(y1)
    
plt.figure(figsize = (14, 8), dpi=300)
plt.plot(x, y, linewidth=4.0, color=(.1, .6, .6))
plt.grid(True)
plt.show()
```

```python
x = list()
y = list()

a = -5
b = 2.2
step = 0.1

for i in range(int((b-a)/step)):
    x1 = a + i * step
    y1 = x1**3 - x1 + exp(-x1)
    
    x.append(x1)
    y.append(y1)
    
plt.figure(figsize = (14, 8), dpi=300)
plt.plot(x, y, linewidth=4.0, color=(.7, .3, .1))
plt.grid(True)
plt.show()
```
