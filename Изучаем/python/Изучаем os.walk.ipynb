os.walk() возвращает генератор:

```python
import os
from pprint import pprint

w = os.walk('./')
print(w)
```

Обращаться к генератору можно с помощью:

- метода next()
- цикла for
- преобразования list()
- передаче методам, которые неявно выполняют преобразование к list
    (например, sorted)


После каждой итерации возвращается кортеж из 3-х элементов:

- текущего каталога
- списка директорий  втекущем каталоге
- списка файлов в текущем каталоге

```python
pprint(next(w))
```

```python
pprint(next(w))
```

```python
pprint(next(w))
```

```python
pprint(next(w))
```

```python
pprint(list(w))
```
