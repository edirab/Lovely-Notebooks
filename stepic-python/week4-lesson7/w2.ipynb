Производятся независимые испытания. В каждом испытании возможно всего два исхода: «успех» и «неудача», причём вероятность успеха равна $p$. Тогда вероятность того, что в $n$ испытаниях произойдёт ровно $k$ успехов, равна $C_{n}^{k}p^{k}(1-p)^{n-k}$

### Задача 1

Пусть вероятность попадания в цель равна 1/5. Производится 10 независимых выстрелов. Какова вероятность попадания в цель по меньшей мере дважды?


#### Подсказка

Рассмотрим обратное событие - его вероятность найти легче.

#### Решение:

Попасть по меньшей мере дважды, эквивалентно 2, 3, 4, 5, 6, 7, 8, 9 и 10 попаданиям. Однако проще рассмотреть это событие как отрицание события попасть 0 или 1 раз.

$P_{2-10} = \overline{P_{0-1}} = 1 - P_{0} - P_{1}$

Вероятность ровно 1 успеха и ровно 0 успехов легко посчитать по формуле Бернулли $C_{n}^{k}p^{k}(1-p)^{n-k}$. Тогда:

$P_{2-10} = 1 - C_{10}^{0}p^{0}(1-p)^{10-0} - C_{10}^{1}p^{1}(1-p)^{10-1}$
, где p = 1/5

$P_{2-10} = 1 - (1-\frac{1}{5})^{10} - 10(\frac{1}{5})^{1}(1-\frac{1}{5})^{9} = 1 - (\frac{4}{5})^{10} - 2(\frac{4}{5})^{9}$

#### Ответ:
$1 - (\frac{4}{5})^{10} - 2(\frac{4}{5})^{9}$ или 0.62 (при округлении до 2 знака)

    а) опыт закончится до шестого бросания;

#### Подсказка 2

Подсказка 1 делает абсолютно неважным какой стороной выпала монета на 1й бросок. 1й бросок вообще не участвует в вычислениях.

#### Подсказка 3а

#### Решение (пункт "а"):

Для завершения опыта достаточно выпадения подряд любой из 2 сторон дважды (как орла, так и решки), т.е. чтобы опыт НЕ прерывался события успеха и неудачи должны чередоваться после каждого броска.
    - Подбросим монетку 1-й раз. Запишем результат (не важно какой).
    - Подбросим монетку 2-й раз. Сравним результат с результатом на предыдущем шаге. Получим "Успех", если результат совпадает с результатом 1 шага. И "неудачу", если не совпадает.
    - Продолжим подбрасывать монетку  до окончания опыта (либо выпадения 2 одинаковых результатов подряд, либо истечения 5 попыток). Будем записывать "Успех" в случае совпадения результата с предыдущим шагом и "Неудачу", в случае несовпадения.
Очевидно, что вероятность успеха или неудачи p на каждом шаге равна 1/2, т.к. монетка уравновешена и на результат не влияют предыдущие испытания.

$p = \frac{1}{2}$

Рассмотрим вероятность 5 неудач:

$P_{5 неуд} = C_{5}^{5}p^{5}(1-p)^{5-5}$

Любой другой результат (хоть 1 удача) - искомое событие. Следовательно его вероятность:

$P_{удачи} = 1 - P_{5 неуд} = 1-(\frac{1}{2})^{5} = \frac{15}{16}$

#### Альтернативное решение (пункт "а"):

Вспомним условную вероятность (1-я неделя).
    - Итак, бросили монету первый раз. Какой-то стороной она упадёт. Это не важно. С первого раза мы процесс не закончим.
    - Во второй раз она может упасть той же стороной, а может и другой. И оба случая равновероятны и от первого раза не зависят.
        - Если той же стороной, то процесс закончен.
        - Если нет, то кидаем ещё раз. Опять монета может упасть той же стороной, что в предыдущий раз (не в первый!). И повторяем до бесконечности. 
Получается, что мы чётко задаём последовательность успехов и неудач:

- успех на 2м броске, $P(+2) = \frac{1}{2}$
- **неуспех** на 2м броске, $P(-2) = \frac{1}{2}$ и успех на 3м, $P(+3) = \frac{1}{2}$. ИТОГО $P(-2,+3) = \frac{1}{4}$

Суммарная вероятность тогда будет иметь вид ряда:

$P = \frac{1}{2} + \frac{1}{4} + \frac{1}{8} + \frac{1}{16} + \frac{1}{32} + ...$

Для 5 испытаний (1-е определяет начальную сторону, а 4 последующих определяют успех/неудачу) получаем:

$P = \frac{1}{2} + \frac{1}{4} + \frac{1}{8} + \frac{1}{16} = \frac{5}{16}$

#### Ответ (пункт "а"):
$\frac{15}{16}$

#### Решение (пункт "б"):

Рассуждаем аналогично пункту а.

Заметим, что чётное число бросков, это: 2, 4, 6, 8 ... При этом 1й бросок всегда "контрольный". Т.е. на самом деле нас интересуют 1, 3, 5, 7 и т.д. испытания по схеме выше.

Что значит, что i-е испытание успешно? Это значит, что ровно (i-1) перед ним не успешны, а i-е успешно.

$P_{i-ого успеха} = p^{1}(1-p)^{i-1}$

Следовательно вероятность успеха на чётном броске, это сумма вероятностей нечётных испытаний (т.к. 1-й бросок "контрольный" и задаёт сторону для успехов/неудач следующих бросков-испытаний):

$P_{успеха чётного броска} = P_{успеха нечётного испытания} = \sum_{i=1}^{\propto}P_{(2i+1)-ого успеха} = \sum_{i=1}^{\propto}(p(1-p)^{2i})$

Отсюда можем представить нашу вероятность как сумму бесконечного ряда:

$P_{успеха чётного броска} = \frac{1}{2} + \frac{1}{8} + \frac{1}{32} + \frac{1}{128} + ... = \frac{1}{2}(1 + \frac{1}{4} + \frac{1}{16} + \frac{1}{64} + ...)$

Заметим, что сумма ряда $\frac{1}{4} + \frac{1}{16} + \frac{1}{64} + ...$ является извесной величиной из <a href ="https://en.wikipedia.org/wiki/1/4_%2B_1/16_%2B_1/64_%2B_1/256_%2B_%E2%8B%AF">Wiki</a> и равна $\frac{1}{3}$, следовательно:

$P_{успеха чётного броска} = \frac{1}{2}(1+ \frac{1}{3}) = \frac{2}{3}$

#### Ответ (пункт "б"):

$\frac{2}{3}$

### Задача 3

Из урны, содержащей 2n белых и 2n черных шаров, наудачу извлекают (с возвращением) 2n шаров. Какова вероятность того, что среди них окажется поровну шаров белого и черного цвета?

#### Решение

Т.к. мы извлекаем всего 2n шаров из которых должно быть поровну белых и чёрных, это означает, что мы должны вытащить ровно n белых и ровно n чёрных.

Т.к. шары возвращаются в урну, то вытащить белый шар всегда $\frac{2n}{2n+2n} = \frac{1}{2}$ (число белых шаров в урне к числу всех шаров).

Примем событие "вытащить белый шар" за успех (желающие могут принять за успех вероятность извлечь чёрный шар, на результат это не повлияет). Таким образом, мы видим классическую задачу на схему Бернулли:


- $N = 2n$ - число испытаний
- $k = n$ - число успехов
- $p = \frac{1}{2}$ - вероятность успеха
- $P = C_{N}^{k}p^{k}(1-p)^{N-k}$

$P = C_{2n}^{n}(\frac{1}{2})^{n}(1-\frac{1}{2})^{2n-n} = \frac{1}{2^{2n}}$

#### Ответ:

$\frac{C_{2n}^{n}}{2^{2n}}$
