### Вычисление факториала

```python
def fact(n):
    if n == 1:
        return 1
    else:
        return (n * fact(n-1))
```

```python
%%timeit
fact(100)
```

### Числа Фибоначчи

```python
def fib(n):
    a = 0
    b = 1
    if n == 1 or n == 2:
        return 1
    
    for i in range(n-1):
        n = a + b
        a = b
        b = n
    return n
```

```python
for i in range(1, 10):
    print(fib(i))
```

### Простые числа

```python
def is_prime(n):
   
    if n % 2 == 0 or n % 5 == 0:
        return False
    
    max_divider = int(n**0.5) + 1
    
    for i in range(2, max_divider):
        if n % i == 0:
            return False
    return True
```

```python
%%timeit
for i in range(1, 10000):
    if is_prime(i):
        print(i, " is prime", )
```

```python
def both_ends(s):
    r = ''
    if len(s) > 1:
        r += s[:2]
        r += s[-2:]
    return r
```

```python
s = 'qwerty'
print(both_ends(s))
```

```python
def fix_start(s):
    r = ''
    if len(s):
        first = s[0]
        r += first
        for i in range(1, len(s)):
            if s[i] == first:
                r += '*'
            else:
                r += s[i]
    return r
```

```python
s = ''
print(fix_start(s))
```

### Факультатив недели №1

```python
def numerics(n):
    return list(map(int, str(n)))
```

```python
numerics(4224)
```

```python
def kaprekar_step(L):
    LL1 = L.copy()
    LL2 = L.copy()

    LL1.sort()
    LL2.sort(reverse=True)
    #print(LL1, LL2)
    
    s1 = int(''.join(map(str, LL1)))
    s2 = int(''.join(map(str, LL2)))
    
    #print(s1, s2, max(s1, s2) - min(s1, s2))
    res1 = (max(s1, s2) - min(s1, s2))
    
    return res1
```

```python
kaprekar_step([4, 5, 6])
#kaprekar_step([7, 3, 3, 9])
```

```python
def kaprekar_loop(n):
    r = n
    
    if r == 1000:
        print("Ошибка! На вход подано число 1000")
        return
    if len(set(numerics(r))) == 1:
        print("Ошибка! На вход подано число {} - все цифры одинаковые".format(r))
        return
    
    print(r)    
    
    if n == 6174:
        return

    while r != 6174:
        r_list = numerics(r)
        r = kaprekar_step(r_list)
        print(r)
    return
```

```python
kaprekar_loop(1000)
```

```python
def kaprekar_check(n):
    s = list(str(n))
    
    if len(s) == 3 or len(s) == 4 or len(s) == 6:
        if len(set(s)) != 1:
            if n != 100 and n != 1000 and n != 100000:
                return True
    return False
```

```python
kaprekar_check(1000)
```

```python
def kaprekar_loop(n):
    
    kap_nums = {495, 6174, 549945, 631764}
    
    if kaprekar_check(n):
        print(n)
        prev_numbers= set()
        ka = n
        
        while (ka not in kap_nums):
            ka_list = numerics(ka)
            ka = kaprekar_step(ka_list)
            if ka in prev_numbers:
                print("Следующее число - {}, кажется процесс зациклился...".format(ka))
                return
            else:
                prev_numbers.add(ka)
                print(ka)
        
        
    else:
        print("Ошибка! На вход подано число {}, не удовлетворяющее условиям процесса Капрекара".format(n))
    return
```
