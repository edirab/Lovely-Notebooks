### Массштаб гистограмм может быть обманчивым...

```python
import numpy as np


fig, ax = plt.subplots(figsize=(15,10))
D[D.bachelor != 0].hist(column='bachelor', ax = ax, bins = 100)
x = np.arange(1, 1600)
def f(x):
    return 0*x+1
def g(x):
    return (800.*x)/(x+25.)**2

ax.plot(x, f(x), lw=5)
ax.plot(x, g(x), lw=5)
```

```python
import numpy as np


fig, ax = plt.subplots(figsize=(15,10))
D[D.bachelor != 0].hist(column='bachelor', ax = ax, bins = 1000)
x = np.arange(1, 1600)

ax.plot(x, f(x), lw=5)
ax.plot(x, g(x), lw=5)
```
