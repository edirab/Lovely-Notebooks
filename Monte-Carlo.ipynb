```python
from matplotlib import pyplot as plt
import numpy as np
from math import sin, cos, pi


# Рисуем квадрат
square_x = [-1,1, 1, -1, -1]
square_y = [-1, -1, 1, 1, -1]


# Рисуем круг внутри квадрата
phi = np.linspace(0, 2*pi)
circle_x = np.cos(phi)
circle_y = np.sin(phi)


n_dots = 220
n_inside = 0

dots_x = np.random.rand(1, n_dots)
dots_y = np.random.rand(1, n_dots)

dots_x = (dots_x - 0.5) * 2
dots_y = (dots_y - 0.5) * 2

radius = dots_x ** 2 + dots_y ** 2

# print((radius < 1)[0][:10])
n_inside = np.count_nonzero(radius < 1)
    

plt.figure(figsize=(14, 14))
plt.plot(square_x, square_y)
plt.plot(circle_x, circle_y)
plt.scatter(dots_x, dots_y, color = 'c')
plt.grid()

plt.show()
print("Точек попало в круг: {:}".format(n_inside))
print("Всего: {:16}".format(n_dots))
print("Число Pi: {}".format( 4 * n_inside / n_dots))
```
