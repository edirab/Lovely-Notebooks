```python
from matplotlib import pyplot as plt
import numpy as np
from math import sin, cos, pi

#  
square_x = [-1,1, 1, -1, -1]
square_y = [-1, -1, 1, 1, -1]

#    
circle_x = []
circle_y = []
phi = np.linspace(0, 2*pi)

for i in phi:
    circle_x.append(cos(i))
    circle_y.append(sin(i))


n_dots = 220
n_inside = 0

#np.random.

dots_x = []
dots_y = []
for i in range(n_dots):
    dot = np.random.rand(1, 2)
    dot = (dot - 0.5) * 2
    
    dots_x.append(dot[0][0])
    dots_y.append(dot[0][1])
    
    if dot[0][0]**2 + dot[0][1]**2 < 1:
        n_inside += 1
    

plt.figure(figsize=(14, 14))
plt.plot(square_x, square_y)
plt.plot(circle_x, circle_y)
plt.scatter(dots_x, dots_y, color = 'c')
plt.grid()

plt.show()
print("   : {:}".format(n_inside))
print(": {:16}".format(n_dots))
print(" Pi: {}".format( 4 * n_inside / n_dots))
```
